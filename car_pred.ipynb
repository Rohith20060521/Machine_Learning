```python
x=df['enginesize']
y=df['price']
```

```python
plt.scatter(x,y)
plt.xlabel('Engine Size')
plt.ylabel('Price')
plt.title('Engine Size vs Price')
plt.show()
```

```python
# Training the Linear Regression model on the whole dataset
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(x.values.reshape(-1,1), y)
from sklearn.preprocessing import PolynomialFeatures
poly = PolynomialFeatures(degree=4)
X_poly =poly.fit_transform(x.values.reshape(-1,1))
poly_reg = LinearRegression()
poly_reg.fit(X_poly, y)
x_sorted=np.sort(x)
```

```python
x_range=np.linspace(min(x), max(x), 100).reshape(-1,1)
X_range_poly = poly.transform(x_range)
plt.plot(x_range,poly_reg.predict(X_range_poly), color='green')
plt.scatter(x, y)
plt.plot(x, lin_reg.predict(x.values.reshape(-1,1)), color='orange')
plt.xlabel('Engine Size')
plt.ylabel('Price')
plt.title('Engine Size vs Price with Regression Lines')
plt.show()
```

```python
line_pred=lin_reg.predict([[200]])
print(f"Linear Regression Prediction for 200: {line_pred}")
poly_pred=poly_reg.predict(poly.fit_transform([[200]]))
print(f"Polynomial Regression Prediction for 200: {poly_pred}")
```
