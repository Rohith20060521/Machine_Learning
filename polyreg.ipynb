```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('D:/5jq/ML/Position_Salaries.csv')
```

```python
x=df.iloc[:,1:-1].values
y=df.iloc[:,-1].values
```

```python
print("Dataset preview:")
print(df.head())
```

```python
# Training the Linear Regression model on the whole dataset
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(x, y)
 
# Training the Polynomial Regression model on the whole dataset
from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures(degree = 4)
X_poly = poly_reg.fit_transform(x)
lin_reg_2 = LinearRegression()
lin_reg_2.fit(X_poly, y)
 
print("✅ Models Trained Successfully!")
```

```python
plt.scatter(x, y, color = 'orange')
plt.plot(x, lin_reg.predict(x), color = 'green')
plt.xlabel('Position level')
plt.ylabel('Salary')
plt.title('Linear Regression')
plt.show()
```

```python
plt.scatter(x, y, color = 'red')
plt.plot(x,lin_reg_2.predict(poly_reg.fit_transform(x)), color = 'orange')
plt.xlabel('Position level')
plt.ylabel('Salary')
plt.title('Polynomial Regression')
plt.show()
```

# Prediction

```python
line_pred=lin_reg.predict([[7.5]])
print(f"Linear Regression Prediction for 7.5: {line_pred}")
poly_pred=lin_reg_2.predict(poly_reg.fit_transform([[7.5]]))
print(f"Polynomial Regression Prediction for 7.5: {poly_pred}")
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np
lin_y_pred = lin_reg.predict(x)
poly_y_pred = lin_reg_2.predict(X_poly)
print("Linear Regression Performance:")
print(f"MAE: {mean_absolute_error(y, lin_y_pred)}")
print(f"MSE: {mean_squared_error(y, lin_y_pred)}")
print(f"RMSE: {np.sqrt(mean_squared_error(y, lin_y_pred))}")
print(f"R²: {r2_score(y, lin_y_pred)}")
print("\nPolynomial Regression Performance:")
print(f"MAE: {mean_absolute_error(y, poly_y_pred)}")
print(f"MSE: {mean_squared_error(y, poly_y_pred)}")
print(f"RMSE: {np.sqrt(mean_squared_error(y, poly_y_pred))}")
print(f"R²: {r2_score(y, poly_y_pred)}")
```
